```python
import requests
import time
import os
COUNT = 300
URL = os.getenv("URL")
JSON_FILE = os.getenv("JF")

with open(JSON_FILE, "rb") as f:
    payload = f.read()

for i in range(1, COUNT + 1):
    print(f"▶ Run {i} / {COUNT}")

    r = requests.post(
        URL,
        data=payload,
        headers={"Content-Type": "application/json"},
        timeout=60
    )

    if r.ok:
        print("✔ Done")
    else:
        print(f"❌ Error {r.status_code}: {r.text}")

    # opsional: kasih jeda biar queue gak jebol
    # time.sleep(0.2)
```
